```python
import re
```

**Size-Based** — split into equal character lengths. Simple, works on anything, but cuts words mid-sentence. Fix this with overlap — each chunk shares some characters with its neighbors.

**Structure-Based** — split on headers/sections (like Markdown `##`). Cleanest results but only works if your documents have consistent structure.

**Semantic-Based** — group sentences by meaning using NLP. Best quality, most expensive and complex to implement.

**Sentence-Based** — split into sentences, group them into chunks with overlap. Good middle ground for most text.

```python
def chunk_by_char(text, chunk_size=150, chunk_overlap=20):
    chunks = []
    start_idx = 0
    while start_idx < len(text):
        end_idx = min(start_idx + chunk_size, len(text))
        chunks.append(text[start_idx:end_idx])
        start_idx = end_idx - chunk_overlap if end_idx < len(text) else len(text)
    return chunks

def chunk_by_section(text):
    return re.split(r"\n## ", text)

def chunk_by_sentence(text, max_sentences=5, overlap=1):
    sentences = re.split(r"(?<=[.!?])\s+", text)
    chunks = []
    start_idx = 0
    while start_idx < len(sentences):
        end_idx = min(start_idx + max_sentences, len(sentences))
        chunks.append(" ".join(sentences[start_idx:end_idx]))
        start_idx += max_sentences - overlap
        if start_idx < 0:
            start_idx = 0
    return chunks
```

```python
sample = "This is sentence one. This is sentence two. Here is another sentence. And one more. Final sentence here."

print(chunk_by_char(sample, chunk_size=100, chunk_overlap=20))
print(chunk_by_sentence(sample, max_sentences=3, overlap=1))
```
